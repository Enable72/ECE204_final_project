```python
import matplotlib.pyplot as plt
```

```python
time_limit = [1,2,4,5,6,8,10]
svmrbf_accuracy = [0.485, 0.91, 0.972, 0.977, 0.974, 0.976, 0.974]
rfn1_accuracy = [0.94, 0.977, 0.959, 0.960, 0.955, 0.952, 0.953]
logistic_accuracy = [0.998, 0.995, 0.996, 0.998, 0.997, 0.9998, 0.9998]
```

```python
plt.plot(time_limit, svmrbf_accuracy, label='SVM rbf')  # Default color and line style
plt.plot(time_limit, logistic_accuracy, 'r--', label='Logistic')  # Red dashed line
plt.plot(time_limit, rfn1_accuracy, 'g:', label='Random Forest 1 estimator')  # Green dotted line

# Optional: add a legend to explain which line is which
plt.legend()
plt.xlabel('Years')
plt.ylabel('Accuracy')
plt.title('NMF methods classifications accuracy comparsion under different year')
# Show the plot
plt.show()
```
